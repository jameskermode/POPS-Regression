```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

$$y(x)=x^3$$
$$y_m(x,\Theta) = \Theta_0x + \Theta_1x^2$$
$$\Theta_{0x} =x^2-\Theta_{1x}x$$



```python

N = 100
X = np.linspace(0.1,4.0,N)

fea = lambda x: np.array([1.,x])
fun = lambda x : x**2

fea = lambda x: np.array([1.,x])*x
fun = lambda x : x**3


F = np.array([fea(x) for x in X])


y = fun(X)
iA = F.T@F
T_m = np.linalg.solve(iA,(F*y[:,None]).sum(0))
y_m = F@T_m

P = []
for i,x in enumerate(X):
    f = fea(x)
    e = fun(x)-f@T_m
    h = f@np.linalg.solve(iA,f)
    P += [e/h * np.linalg.solve(iA,f) + T_m]

P = np.array(P)
bounds = [(F@(P.T)).min(1),(F@(P.T)).max(1)]



for i,x in enumerate(X):
    fig, axs = plt.subplots(1,2,figsize=(6,3),dpi=200)
    
    f = fea(x)
    e = fun(x)-f@T_m
    h = f@np.linalg.solve(iA,f)
    T_x = e/h * np.linalg.solve(iA,f) + T_m
    T_1 = np.linspace(2.,6.,201)
    T_0 = (fun(x) - f[1]*T_1)/f[0]
    axs[0].set_xlim(-.5,4.5)
    axs[0].set_ylim(bounds[0].min(),bounds[1].max())
    axs[1].set_xlim(P[:,0].min()-1.0,P[:,0].max()+1.0)
    axs[1].set_ylim(P[:,1].min()-1.0,P[:,1].max()+1.0)
    
    axs[0].fill_between(X,bounds[0],bounds[1],facecolor='C0',alpha=0.5,label='POPS max/min')
    axs[0].plot(X,F@T_x,label=r'POPS $y=\hat{f}(x,{\Theta}^*_{%2.2g})$' % x)
    axs[0].plot(X,F@T_m,'--',c='k',lw=2,label=r'Min loss $y=\hat{f}(x,\bar{\Theta})$')
    axs[0].plot(X,y,'k-',lw=2,label=r'Truth $y=f(x)$')
    axs[0].plot(x,fun(x),'C0o')
    
    axs[1].plot(P[:,0],P[:,1],'-',lw=4,c='0.5',label='POPS ensemble')

    axs[1].plot(T_0,T_1,c='C0',label=r'POPS $\mathcal{P}_\mathcal{M}(%2.2g)$' % x)
    axs[1].plot(T_x[0],T_x[1],'C0o',label=r'POPS fit $\Theta^*_X$')#,c=cm.viridis(x/X.max()))
    axs[1].plot(T_m[0],T_m[1],'ks',label=r'Min loss fit $\bar{\Theta}$')

    axs[1].set_xlabel(r"Parameter $\Theta_0$",fontsize=8)
    axs[0].set_xlabel(r"Input $x$",fontsize=8)
    axs[1].set_ylabel(r"Parameter $\Theta_1$",fontsize=8)
    axs[0].set_ylabel(r"Output $y$",fontsize=8)
    
    
    axs[1].legend(fontsize=8,loc='upper right')
    axs[0].legend(fontsize=8,loc='upper left')
                
    plt.tight_layout()
    #plt.show()
    plt.savefig(f"img_{i}.png")
    plt.close()
#ax.legend(fontsize=8)
#sm = plt.cm.ScalarMappable(cmap=cm.viridis)
#sm.set_array([])
#fig.colorbar(sm, ax=ax, orientation='horizontal', label='x')

```
